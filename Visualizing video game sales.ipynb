# This is a demonstration of my skills as a data analyst using an example dataset of video game sales from 1984 - 2014. 

This dataset contains a list of video games with sales greater than 100,000 copies. It was generated by a scrape of vgchartz.com.

sourceurl: https://www.kaggle.com/code/upadorprofzs/eda-video-game-sales

I will clean the dataset and manipulate the data to get my goal of showcasing the top 30 selling North American Video Games for this time period

Tableau visual: https://public.tableau.com/app/profile/matt.shuff/viz/top25vgsalesinNA/Sheet1?publish=yes

```python
#importing libraries

import pandas as pd

import numpy as np
```

```python
#read in datafile and convert to pandas dataframe

csv = pd.read_csv('vg.csv')

df = pd.DataFrame(csv)
```

```python
df
```

```python
#truncating list down to top 30 selling North American video games

df_trunc = df.sort_values('NA_Sales', ascending=False).head(30)

df
```

```python
#locating nan values for removal

df.Year_of_Release.value_counts(dropna=False)
```

```python
#Dropping rows of data with nan values for the year of release

df = df.dropna(subset = 'Year_of_Release')
```

```python
#converting year of release column to integer for formatting

df.Year_of_Release = df.Year_of_Release.astype(int)
```

```python
df.sort_values('Year_of_Release', ascending=False)
```

```python
#Sorting by top 25 best selling games 

df_na = df.sort_values('NA_Sales',ascending = False).head(25)
```

```python
#converting year of release column to str to add into chart title

df['Year_of_Release'] = df['Year_of_Release'].astype(str)
```

```python
df_na['Name_YOR'] = df_na['Name'] + ' ' + df['Year_of_Release']
```

```python
df_na
```

```python
df_na.to_csv('top 25 NA video games by Year.csv')
```
